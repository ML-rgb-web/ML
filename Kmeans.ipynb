```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, ConfusionMatrixDisplay

data = pd.read_csv('brain_tumor_dataset.csv')


data = data.dropna(
    subset=[
        'Age',
        'Tumor_Size',
        'Tumor_Growth_Rate',
        'Survival_Rate',
        'Tumor_Type'
    ]
)

X = data[
    ['Age',
     'Tumor_Size',
     'Tumor_Growth_Rate',
     'Survival_Rate']
]

y_true = data['Tumor_Type'].map({
    'Benign': 0,
    'Malignant': 1
}).values

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)


model = KMeans(
    n_clusters=2,
    random_state=42,
    n_init='auto'
)


clusters = model.fit_predict(X_scaled)


if (clusters == y_true).mean() < 0.5:
    clusters = 1 - clusters


print(classification_report(y_true, clusters))


ConfusionMatrixDisplay.from_predictions(
    y_true,
    clusters,
    cmap='Blues'
)

plt.title("KMeans Confusion Matrix")

plt.show()
```
